```python
# 添加图例和标题

import matplotlib.pyplot as plt
import numpy as np


plt.rcParams["font.sans-serif"] = "SimHei"
plt.rcParams["axes.unicode_minus"] = False


x = np.linspace(-2*np.pi, 2*np.pi, 200)
y = np.sin(x)
y1 = np.cos(x)


plt.title("正弦函数和余弦函数的折线图")
plt.plot(x, y, label=r"$\sin(x)$")
plt.plot(x, y1, label=r"$\cos(x)$")
plt.legend(loc="lower left")
plt.show()
```

```python
# 图例的展示样式调整

x = np.arange(0, 2.1, 0.1)
y = np.power(x, 3)
y1 = np.power(x, 2)
y2 = np.power(x, 1)


plt.plot(x, y, ls="-", lw=2, label="$x^3$")
plt.plot(x, y1, ls="-", lw=2, c="r", label="$x^2$")
plt.plot(x, y2, ls="-", lw=2, c="y", label="$x^1$")
plt.legend(loc="upper left", bbox_to_anchor=(0.05, 0.95), ncol=3,
           title="power function", shadow=True, fancybox=True)
plt.show()
```

```python
# 标题的展示样式调整


x = np.linspace(-2, 2, 1000)
y = np.exp(x)


plt.plot(x, y, ls="-", lw=2, color="g")
plt.title("center demo")
plt.title("Left Demo", loc="left", fontdict={
          "size": 12, "color": "r", "family": "Times New Roman"})
plt.title("right demo", loc="right", family="Comic Sans MS",
          size=20, style="oblique", color="c")
plt.show()
```

```python
# 调整刻度范围和刻度标签


x = np.linspace(-2*np.pi, 2*np.pi, 200)
y = np.sin(x)


plt.subplot(211)
plt.plot(x, y)
plt.subplot(212)
plt.xlim(-2*np.pi, 2*np.pi)
plt.xticks([-2*np.pi, -3*np.pi/2, -1*(np.pi)/2, 0, (np.pi)/2, np.pi, 3*np.pi/2, 2*np.pi],
           ["$-3\pi/2$", "$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$3\pi/2$", "$2\pi$", "$2\pi$"])
plt.plot(x, y)
plt.show()
```

```python
# 逆序设置坐标轴刻度标签


time = np.arange(1, 11, 0.5)
machinePower = np.power(time, 2)+0.7


plt.plot(time, machinePower, linestyle="-", linewidth=2, color="r")
plt.xlim(10, 1)
plt.xlabel("使用年限")
plt.ylabel("机器功率")
plt.title("机器损耗曲线")
plt.grid(ls=":", lw=1, color="gray", alpha=0.5)
plt.show()
```

```python
# 向统计图添加表格

# 绘图
labels = "A难度水平", "B难度水平", "C难度水平", "D难度水平"
students = [0.35, 0.15, 0.20, 0.30]
explode = [0.05]*4

plt.title("选择不同难度测试试卷的学生占比")
plt.pie(students, explode=explode, labels=labels, autopct="%1.1f%%",
        startangle=45, shadow=True)


# 制表
colLabels = ["语文", "数学", "英语", "物理"]
rowLabels = ["alice"]
Values = [[112, 130, 98, 88]]
colColors = ["r", "g", "b", "y"]

plt.table(cellText=Values,
          cellLoc="center",
          colWidths=[0.3]*4,
          colLabels=colLabels,
          colColours=colColors,
          rowLabels=rowLabels,
          rowLoc="center",
          loc="bottom")
plt.show()
```
