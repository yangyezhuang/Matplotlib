```python
# plot()函数

import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'simhei'
plt.rcParams['axes.unicode_minus'] = False

# 函数xlabel()和ylabel()——设置x轴和y轴的标签文本

x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")

# 设置标签
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("y=sin(x)")# 函数title()——添加图形内容的标题
plt.legend(loc="lower left")# 函数legend()——标识不同图形的文本标签图例

plt.show()
```

```python
# 函数grid()——绘制刻度线的网格线

x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()
plt.grid(ls='-.', c='r', alpha=0.5)
plt.show()
```

```python
# 函数axhline()、axvline()——绘制平行于x/y轴的水平参考线

x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()

plt.axhline(y=0.0, c='r', ls='--', lw=2)
plt.axvline(x=4, c='r', ls='--', lw=2)

plt.show()
```

```python
# 函数axvspan()——绘制垂直于x轴的参考区域

x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()

plt.axhspan(ymin=0.0, ymax=0.5, facecolor='y', alpha=0.3)
plt.axvspan(xmin=4.0, xmax=6.0, facecolor='y', alpha=0.3)

plt.show()
```

```python
# 函数annotate()——添加图形内容细节的指向型注释文本

x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()
plt.annotate("maximum",
             xy=(np.pi/2, 1.0),
             xytext=((np.pi/2)+1.0, .8),
             weight="bold",
             color="b",
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))

plt.show()
```

```python
# 函数text()——添加图形内容细节的无指向型注释文本

x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()

plt.text(3.1, 0.09, "y=sin(x)", color="b")

plt.show()
```

```python
# scatter()函数

x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x, y, label='scatter figure')
plt.legend()
# 函数xlim()与ylim()——设置x轴与y轴的数值显示范围
plt.xlim(0, 10)
plt.ylim(0, 1)
plt.show()
```

```python
# 函数bar()——用于绘制柱状图

x = ["q", "a", "c", "e", "r", "j", "b", "p"]
y = [3, 1, 4, 5, 8, 9, 7, 2]

plt.bar(x, y,  color="c", hatch="/")
plt.xlabel('box num')
plt.ylabel('weight(kg)')
plt.show()
```

```python
# 函数hist()——用于绘制直方图

boxWeight = np.random.randint(0, 10, 100)
x = boxWeight
bins = range(0, 11, 1)

plt.hist(x, bins=bins, color="g", histtype="bar", rwidth=1, alpha=0.8)
plt.xlabel("box weight(kg)")
plt.ylabel("sales count")
plt.show()
```

```python
# 函数pie()——用于绘制饼图

kinds = ["简易箱", "保温箱", "行李箱", "密封箱"]
soldNums = [0.05, 0.45, 0.15, 0.35]

plt.pie(soldNums, labels=kinds, autopct='%.2f%%')
plt.title('不同类型箱子的销售比例')
plt.show()
```

```python
# 函数polar()——用于绘制极线图

barSlices = 12
theta = np.linspace(0, 2*np.pi, barSlices, endpoint=False)
r = 30*np.random.rand(barSlices)

plt.title('极线图')
plt.polar(theta, r, color="chartreuse",
          linewidth=2, marker="*", mfc="b", ms=10)
plt.show()
```

```python
# 函数polar()——用于绘制极线图

kinds = ["语文", "数学", "英语", "化学","历史","地理"]
soldNums = [120, 130, 95, 115,112,99]
theta = np.linspace(0, 2*np.pi, len(kinds), endpoint=False)
theta = np.concatenate((theta, [theta[0]]))
kinds = np.concatenate((kinds, [kinds[0]]))
soldNums = np.concatenate((soldNums, [soldNums[0]]))

plt.polar(theta, soldNums,'bo-')
plt.fill(theta, soldNums,alpha=0.4)
plt.thetagrids(theta*180/np.pi, kinds)
plt.show()
```

```python
# 函数scatter()——用于绘制气泡图


a = np.random.randn(100)
b = np.random.randn(100)

plt.scatter(a, b, s=np.power(10*a + 20*b, 2),
            c=np.random.rand(100),
            cmap=plt.cm.RdYlBu,
            marker="o")
plt.show()
```

```python
# 函数stem()——用于绘制棉棒图

x = np.linspace(0.5, 2*np.pi, 20)
y = np.random.randn(20)

plt.stem(x, y, linefmt="-.", markerfmt="o", basefmt="-")
plt.show()
```

```python
# 函数boxplot()——用于绘制箱线图

x = np.random.randn(100)

plt.boxplot(x)
plt.xticks([1], ["随机数生成器AlphaRM"])
plt.ylabel("随机数值")
plt.title("随机数生成器抗干扰能力的稳定性")
plt.grid(axis="y", ls=":", lw=1, color="gray", alpha=0.4)
plt.show()
```

```python
# 函数errorbar()——用于绘制误差棒图

x = np.linspace(0.1, 0.6, 6)
y = np.exp(x)

plt.errorbar(x, y, fmt="bo:", yerr=0.2, xerr=0.02)
plt.xlim(0, 0.7)
plt.show()
```
