```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'simhei'
plt.rcParams['axes.unicode_minus'] = 'false'

subject = ['语文', '数学', '英语', '物理', '化学']
v1 = [77, 92, 83, 74, 90]
v2 = [63, 88, 99, 69, 66]
```

```python
plt.title('折线图', fontsize=15)
plt.plot(subject, v1, 'o-', label='v1')
plt.plot(subject, v2, 'd-', label='v2')
plt.grid(axis='y', alpha=0.4)
plt.legend()
plt.show()
```

```python
plt.title('')
plt.plot(subject, v1, 'r-', label='v1')
plt.legend()

ax2 = plt.twinx()
ax2.plot(subject, v2, label='v2')
plt.show()
```

```python
plt.bar(np.arange(len(v1)), v1, color='tomato', width=0.4, label='v1')
for x, y in enumerate(v1):
    plt.text(x, y, y, ha='center', va='bottom')

plt.bar(np.arange(len(v2)) + 0.4, v2, width=0.4, label='v2')
for x, y in enumerate(v2):
    plt.text(x + 0.4, y, y, ha='center', va='bottom')

plt.xticks(np.arange(len(v1)) + 0.4 / 2, subject)
plt.legend()
plt.show()
```

```python
plt.bar(np.arange(len(v1)), v1, width=0.6, color='tomato', label='v1')
for x, y in enumerate(v1):
    plt.text(x, y - 20, y, ha='center', va='top')

plt.bar(np.arange(len(v2)), v2, width=0.6, bottom=v1, label='v2')
for x, y in enumerate(v2):
    plt.text(x, y + 60, y, ha='center', va='bottom')

plt.xticks(np.arange(len(v1)), subject)
plt.legend()
plt.show()
```

```python
plt.barh(np.arange(len(v1)), v1, height=0.4, label='v1')
for x, y in enumerate(v1):
    plt.text(y, x, y)

plt.barh(np.arange(len(v2)) + 0.4, v2, height=0.4, label='v2')
for x, y in enumerate(v2):
    plt.text(y, x + 0.4, y)

plt.yticks(np.arange(len(v1)) + 0.4 / 2, subject)
plt.legend()
plt.show()
```

```python
plt.barh(np.arange(len(v1)), v1, height=0.5, label='v1')
for x, y in enumerate(v1):
    plt.text(y, x, y)

plt.barh(np.arange(len(v2)), v2, height=0.5, left=v1, label='v2')
for x, y in enumerate(v2):
    plt.text(y, x, y)

plt.yticks(np.arange(len(v1)), subject)
plt.legend(bbox_to_anchor=[0.7,0,0.5,1])
plt.show()
```

```python
err_value=[5,7,9,6,4]
err_style=dict(elinewidth=2,capsize=3)
plt.bar(subject,v1,yerr=err_value,error_kw=err_style,label='v1')
plt.show()
```

```python
explode=[0.01]*len(v1)
plt.pie(v1,explode,subject,autopct='%.2f%%')
plt.legend(subject,bbox_to_anchor=[-0.7,0,0.5,1])
plt.show()
```

```python
plt.pie(v1,autopct='%.2f%%',pctdistance=0.8,wedgeprops=dict(edgecolor='w',width=0.4))
plt.legend(subject,bbox_to_anchor=[-0.7,0,0.5,1])
plt.show()
```

```python
colormap=['c','pink','skyblue','yellow','green']
plt.pie(v1,autopct='%.1f%%',pctdistance=0.85,colors=colormap,wedgeprops=dict(width=0.3,edgecolor='w'))

plt.pie(v2,autopct='%.1f%%',radius=0.7,pctdistance=0.7,colors=colormap,wedgeprops=dict(width=0.3,edgecolor='w'))

plt.legend(subject,bbox_to_anchor=[-0.7,0,0.5,1])
plt.show()
```

```python
subject = ['语文', '数学', '英语', '物理', '化学']
v1 = [77, 92, 83, 74, 90]
v2 = [63, 88, 99, 69, 66]

subject=np.concatenate((subject,[subject[0]]))
v1=np.concatenate((v1,[v1[0]]))
v2=np.concatenate((v2,[v2[0]]))

theta = np.linspace(0,2*np.pi,len(subject),endpoint=False)

plt.polar(theta,v1)
plt.fill(theta,v1,alpha=0.4,label='v1')

plt.polar(theta,v2)
plt.fill(theta,v2,alpha=0.4,label='v2')

plt.thetagrids(theta*180/np.pi,subject)
plt.legend(bbox_to_anchor=[-0.7,0,0.5,1])
plt.show()
```
