```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'simhei'
plt.rcParams['axes.unicode_minus'] = False

subject = ['语文', '数学', '英语', '物理', '化学']
v1 = [77, 92, 83, 74, 90]
v2 = [63, 88, 99, 69, 66]
```

```python
class Project:
    # 绘图部分
    def graph(self):
        labs = "A难度水平", "B难度水平", "C难度水平", "D难度水平"
        students = [0.35, 0.15, 0.20, 0.30]
        explode = [0.05] * 4
        plt.title("难度等级占比")
        plt.pie(students, explode, labs, autopct="%.1f%%", shadow=True)

    # 表格部分
    def table(self):
        colLabels = ["语文", "数学", "英语", "物理"]
        rowLabels = ["alice", 'tom']
        Values = [[112, 130, 98, 88],
                  [101, 122, 99, 95]]

        colColors = ["r", "g", "b", "y"]
        plt.table(cellText=Values,  # 表格数据
                  cellLoc="center",  # 数据左右对齐方式
                  colWidths=[0.3] * 4,
                  colLabels=colLabels,  # 行标签
                  colColours=colColors,
                  rowLabels=rowLabels,  # 列标签
                  rowLoc="center")  # 数据上下对齐方式


def main():
    p = Project()
    p.graph()
    p.table()
    plt.show()


if __name__ == '__main__':
    main()
```

```python
subject = ['语文', '数学', '英语', '物理', '化学']
v1 = [77, 92, 83, 74, 90]
v2 = [63, 88, 99, 69, 66]

theta = np.linspace(0, 2 * np.pi, len(subject), endpoint=False)
theta = np.concatenate((theta, [theta[0]]))
subject = np.concatenate((subject, [subject[0]]))
v1 = np.concatenate((v1, [v1[0]]))
v2 = np.concatenate((v2, [v2[0]]))

plt.polar(theta, v1)
plt.fill(theta, v1, alpha=0.4, label='v1')
plt.polar(theta, v2)
plt.fill(theta, v2, alpha=0.4, label='v2')
plt.thetagrids(theta * 180 / np.pi, subject)
plt.legend(bbox_to_anchor=(-0.7, 0, 0.5, 1))
plt.title('雷达图')
plt.show()
```

```python
theta = np.linspace(0, 2 * np.pi, len(v1), endpoint=False)

plt.polar()
plt.bar(theta, v1,
        width=1.05,
        bottom=30,
        color=np.random.random((len(v1), 3))
        )

for x, y in zip(theta, v1):
    plt.text(x, y + 40, y)
plt.axis(False)
plt.title('南丁格尔玫瑰图')
plt.show()
```

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, label="sin(x)")

# 函数axhline()、axvline()绘制平行于x/y轴的水平参考线
plt.title('绘制平行于x/y轴的水平参考线')
plt.axhline(y=0.0, c='r', ls='--', lw=2)
plt.axvline(x=4, c='r', ls='--', lw=2)

plt.legend()
plt.show()
```

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2,  label="sin(x)")


# 函数axvspan() 绘制垂直于x轴的参考区域
plt.title('绘制垂直于x轴的参考区域')
plt.axhspan(ymin=0.0, ymax=0.5, facecolor='y', alpha=0.3)
plt.axvspan(xmin=4.0, xmax=6.0, facecolor='y', alpha=0.3)

plt.legend()
plt.show()
```

```python
x = np.linspace(0.1, 0.6, 10)
y = np.exp(x)
error = 0.05 + 0.15 * x
lower_error = error
upper_error = 0. * error
error_limit = [lower_error, upper_error]

plt.title('误差棒图')
plt.errorbar(x, y, yerr=error_limit, fmt=":o", ecolor="y",
             elinewidth=4, ms=5, mfc="c", mec="r", capthick=1, capsize=2)

plt.xlim(0, 0.7)
plt.show()

'''
yerr：单一数值的非对称形式误差范围
fmt：数据点的标记样式和数据点标记的连接线样式
ecolor：误差棒的颜色
elinewidth：误差棒的线条粗细
ms：数据点的大小
mfc，mec：数据点的标记颜色，数据点的标记边缘颜色
capthick：误差棒边界横杠的厚度
capsize：误差棒边界横杠的大小
'''
```
