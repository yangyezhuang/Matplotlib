```python
# 标题样式调整
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.family"] = "SimHei"
plt.rcParams["axes.unicode_minus"] = False

font_dict = {
    "size": 12,
    "color": "r",
    "family": "Times New Roman"
}

x = np.linspace(-2, 2, 1000)
y = np.exp(x)

plt.plot(x, y, ls="-", lw=2, color="g")
plt.title("center demo")
plt.title("Left Demo", loc="left", fontdict=font_dict)
plt.title("right demo", loc="right", family="Comic Sans MS",
          size=20, style="oblique", color="c")
plt.show()
```

```python
# 图例样式调整

x = np.arange(0, 2.1, 0.1)
y = np.power(x, 3)
y1 = np.power(x, 2)
y2 = np.power(x, 1)

plt.plot(x, y, ls="-", lw=2, label="$x^3$")
plt.plot(x, y1, ls="-", lw=2, c="r", label="$x^2$")
plt.plot(x, y2, ls="-", lw=2, c="y", label="$x^1$")
plt.legend(loc="upper left", bbox_to_anchor=(0.05, 0.95), ncol=3,
           title="power function", shadow=True, fancybox=True)
plt.show()
```

```python
# 添加图例,标题,标签

title_font = {
    'title': {
        'color': 'red',
        'size': 16
    },
    'x': {
        'color': 'green',
        'size': 13
    },
    'y': {
        'color': 'blue',
        'size': 13
    }
}

x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y = np.sin(x)
y1 = np.cos(x)

plt.title("正弦函数和余弦函数的图像", fontdict=title_font['title'])
plt.plot(x, y, label=r"$\sin(x)$")
plt.plot(x, y1, label=r"$\cos(x)$")
plt.legend(loc="lower left")
plt.xlabel('x', fontdict=title_font['x'])
plt.ylabel('y', fontdict=title_font['y'])
plt.show()
```
